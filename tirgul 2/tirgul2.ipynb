```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import json
from google.colab import drive
drive.mount('/content/drive')

with open('/content/drive/MyDrive/students.json', 'r') as f:
  data = json.load(f)

students = data['students']
```

```python
student_names = [f"{s['first_name']} {s['last_name']}" for s in students]
dropdown = widgets.Dropdown( options=student_names, description='סטודנט:', layout=widgets.Layout(width='300px') )

txt_name = widgets.Text(description='שם מלא:', layout=widgets.Layout(width='400px'))
txt_email = widgets.Text(description='מייל:', layout=widgets.Layout(width='400px'))
txt_courses = widgets.Textarea(description='קורסים:', layout=widgets.Layout(width='400px', height='100px'))
txt_link = widgets.Text(description='קישור:', layout=widgets.Layout(width='400px'))
txt_show = widgets.Text(description='תוכנית אהובה:', layout=widgets.Layout(width='400px'))

btn_update = widgets.Button( description='עדכן תוכנית אהובה', button_style='success', layout=widgets.Layout(width='200px') )
```

```python
def on_student_change(change):
  selected_name = change['new']
  student = next(s for s in students if f"{s['first_name']} {s['last_name']}" == selected_name)
  txt_name.value = f"{student['first_name']} {student['last_name']}"
  txt_email.value = student['email']
  txt_courses.value = '\n'.join(student['courses'])
  txt_link.value = student['favorite_link']
  txt_show.value = student['favorite_show']
```

```python
def on_update_click(b):

  selected_name = dropdown.value
  # עדכון בזיכרון
  for student in students:
    if f"{student['first_name']} {student['last_name']}" == selected_name:
      student['favorite_show'] = txt_show.value
      break
   # שמירה חזרה לקובץ
  with open('/content/drive/MyDrive/students.json', 'w') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
  print(f"✅ התוכנית האהובה של {selected_name} עודכנה בהצלחה!")
```

```python
dropdown.observe(on_student_change, names='value')
btn_update.on_click(on_update_click)
```

```python
on_student_change({'new': student_names[0]})
```

```python
display(widgets.VBox([ widgets.HTML("<h2 style='text-align:right'>📋 טופס סטודנטים</h2>"), dropdown, txt_name, txt_email, txt_courses, txt_link, txt_show, btn_update ]))
```
